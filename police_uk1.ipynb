```python
import os
import pandas as pd
```

```python
# Path to directory containing CSV files
directory = ('C:\\Users\\jorda\\Desktop\\portfolio_projects\\data_sets\\uk_crime\\2024\\')
```

```python
# Define keywords
keywords = ['-outcomes', '-street', '-stop-and-search']
```

```python
csv_files = [file for file in os.listdir(directory) if file.endswith('.csv')]
```

```python
# Initialize lists to store DataFrames for each category
outcomes_df = []
street_df = []
stop_and_search_df = []
```

```python
# Iterate through CSV files
for file in os.listdir(directory):
    if file.endswith('.csv'):
        print("Reading", file)
        df = pd.read_csv(os.path.join(directory, file))
        if any(keyword in file for keyword in keywords):
            if '-outcomes' in file:
                outcomes_df.append(df)
            if '-street' in file:
                street_df.append(df)
            if '-stop-and-search' in file:
                stop_and_search_df.append(df)
```

```python
# Concatenate DataFrames for each category
if outcomes_df:
    combined_outcomes_df = pd.concat(outcomes_df, ignore_index=True)
    combined_outcomes_df.to_csv('C:\\Users\\jorda\\Desktop\\portfolio_projects\\data_sets\\uk_crime\\2024crime_outcomes.csv', index=False)
    print("Saved outcomes data")

if street_df:
    combined_street_df = pd.concat(street_df, ignore_index=True)
    combined_street_df.to_csv('C:\\Users\\jorda\\Desktop\\portfolio_projects\\data_sets\\uk_crime\\2024crime_street.csv', index=False)
    print("Saved street data")

if stop_and_search_df:
    combined_stop_and_search_df = pd.concat(stop_and_search_df, ignore_index=True)
    combined_stop_and_search_df.to_csv('C:\\Users\\jorda\\Desktop\\portfolio_projects\\data_sets\\uk_crime\\2024crime_stop_and_search.csv', index=False)
    print("Saved stop-and-search data")
```
